##**Library import**##

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt # if you use VS, Pycharm, you just use "imshow()".
print(cv2.__version__)
```

##**Define Function**##

```python
def Lightness_histogram(channel, title, color):
  hist = cv2.calcHist([channel],[0], None,[256],[0,256])
  plt.plot(hist, color=color)
  plt.title(title)
  plt.xlabel("Pixel Value")
  plt.ylabel("Frequency")

def square_crop(image,size):
  height, width = image.shape[:2]
  startX = (int)(width/2)-size
  endX=(int)(width/2)+size
  startY=(int)(height/2)-size
  endY=(int)(height/2)+size
  return image[startY:endY, startX:endX]

def img_procssing(image, colormodel):
  gamma = 0.5
  inv_gamma = 1.0 / gamma
  table = np.array([(i/255.0)**inv_gamma*255 for i in range(256)]).astype("uint8")
  gamma_corrected = cv2.LUT(image, table)

  if colormodel == "HSV":
    hsv = cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl_v = clahe.apply(v)
    merged_hsv = cv2.merge((h, s, cl_v))
    cvt_hsv = cv2.cvtColor(merged_hsv, cv2.COLOR_HSV2BGR)
    return cvt_hsv

  elif colormodel =="HSL":
    hsl = cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2HLS)
    h, l, s = cv2.split(hsl)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl_l = clahe.apply(l)
    merged_hsl = cv2.merge((h, cl_l, s))
    cvt_hsl = cv2.cvtColor(merged_hsl, cv2.COLOR_HLS2BGR)
    return cvt_hsl

  elif colormodel =="LAB":
    lab = cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl_l = clahe.apply(l)
    merged_lab = cv2.merge((cl_l, a, b))
    cvt_lab = cv2.cvtColor(merged_lab, cv2.COLOR_LAB2BGR)
    return cvt_lab

  else:
    yuv = cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2YUV)
    y, u, v = cv2.split(yuv)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl_y = clahe.apply(y)
    merged_yuv = cv2.merge((cl_y, u, v))
    cvt_yuv = cv2.cvtColor(merged_yuv, cv2.COLOR_YUV2BGR)
    return cvt_yuv
```

##**File Upload**##

```python
uploaded = files.upload()

filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # cvt  opencv(BGR) -> plt(RGB)

filename = list(uploaded.keys())[1]
ndimg = cv2.imread(filename)
ndimg = cv2.cvtColor(ndimg, cv2.COLOR_BGR2RGB)
```

##**Imgae Crop**##

```python
damage=square_crop(img,200)
nodamage=square_crop(ndimg,200)

fig, axes = plt.subplots(1, 2, figsize=(20, 5))

axes[0].imshow(damage)
axes[0].set_title("Damage")
axes[0].axis("off")

axes[1].imshow(nodamage)
axes[1].set_title("No Damage")
axes[1].axis("off")
```

##**Convert Colormodel**##

```python
#[Height, Width, Channel]
hsv = cv2.cvtColor(damage, cv2.COLOR_BGR2HSV)
chV_hsv = hsv[:, :, 2]

hsl = cv2.cvtColor(damage, cv2.COLOR_BGR2HLS)
chL_hsl = hsl[:,:,1]

lab = cv2.cvtColor(damage, cv2.COLOR_BGR2LAB)
chL_lab = lab[:,:,0]

yuv = cv2.cvtColor(damage, cv2.COLOR_BGR2YUV)
chY_yuv = yuv[:,:,0]
```

```python
#[Height, Width, Channel]
ND_hsv = cv2.cvtColor(nodamage, cv2.COLOR_BGR2HSV)
ND_chV_hsv = ND_hsv[:, :, 2]

ND_hsl = cv2.cvtColor(nodamage, cv2.COLOR_BGR2HLS)
ND_chL_hsl = ND_hsl[:,:,1]

ND_lab = cv2.cvtColor(nodamage, cv2.COLOR_BGR2LAB)
ND_chL_lab = ND_lab[:,:,0]

ND_yuv = cv2.cvtColor(nodamage, cv2.COLOR_BGR2YUV)
ND_chY_yuv = ND_yuv[:,:,0]
```

##**Lightness Channel Visualization**##

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 5))

axes[0][0].imshow(chV_hsv)
axes[0][0].set_title("V channel")
axes[0][0].axis("off")

axes[0][1].imshow(chL_hsl)
axes[0][1].set_title("L channel(HSL)")
axes[0][1].axis("off")

axes[0][2].imshow(chL_lab)
axes[0][2].set_title("L_channel(LAB)")
axes[0][2].axis("off")

axes[0][3].imshow(chY_yuv)
axes[0][3].set_title("Y_channel")
axes[0][3].axis("off")

axes[1][0].imshow(ND_chV_hsv)
axes[1][0].set_title("V channel")
axes[1][0].axis("off")

axes[1][1].imshow(ND_chL_hsl)
axes[1][1].set_title("L channel(HSL)")
axes[1][1].axis("off")

axes[1][2].imshow(ND_chL_lab)
axes[1][2].set_title("L_channel(LAB)")
axes[1][2].axis("off")

axes[1][3].imshow(ND_chY_yuv)
axes[1][3].set_title("Y_channel")
axes[1][3].axis("off")

plt.show()
```

##**Histogram**##

```python
plt.figure(figsize=(20,10))

plt.subplot(2,4,1)
Lightness_histogram(chV_hsv, "V channel", 'r')
plt.subplot(2,4,2)
Lightness_histogram(chL_hsl, "L channel(HSL)", 'r')
plt.subplot(2,4,3)
Lightness_histogram(lab, "L_channel(LAB)", 'r')
plt.subplot(2,4,4)
Lightness_histogram(chY_yuv, "Y_channel", 'r')

plt.subplot(2,4,5)
Lightness_histogram(ND_chV_hsv, "V channel", 'b')
plt.subplot(2,4,6)
Lightness_histogram(ND_chL_hsl, "L channel(HSL)", 'b')
plt.subplot(2,4,7)
Lightness_histogram(ND_chL_lab, "L_channel(LAB)", 'b')
plt.subplot(2,4,8)
Lightness_histogram(ND_chY_yuv, "Y_channel", 'b')
# after, modify the range
```

##**Image Procssing(No Gamma)**##

```python
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl_chV_hsv = clahe.apply(chV_hsv)
cl_chL_hsl = clahe.apply(chL_hsl)
cl_chL_lab = clahe.apply(chL_lab)
cl_chY_yuv = clahe.apply(chY_yuv)

merged_hsv = cv2.merge((cl_chV_hsv, hsv[:, :, 0],hsv[:, :, 1]))
cvt_hsv = cv2.cvtColor(merged_hsv, cv2.COLOR_LAB2BGR)

merged_hsl = cv2.merge((cl_chL_hsl, hsl[:, :, 0],hsl[:, :, 2]))
cvt_hsl = cv2.cvtColor(merged_hsl, cv2.COLOR_LAB2BGR)

merged_lab = cv2.merge((cl_chL_lab, lab[:, :, 1],lab[:, :, 2]))
cvt_lab = cv2.cvtColor(merged_lab, cv2.COLOR_LAB2BGR)

merged_yuv = cv2.merge((cl_chY_yuv, yuv[:, :, 1],yuv[:, :, 2]))
cvt_yuv = cv2.cvtColor(merged_yuv, cv2.COLOR_LAB2BGR)

fig, axes = plt.subplots(2, 4, figsize=(20, 5))

axes[0][0].imshow(cl_chV_hsv)
axes[0][0].set_title("V channel")
axes[0][0].axis("off")

axes[0][1].imshow(cl_chL_hsl)
axes[0][1].set_title("L channel(HSL)")
axes[0][1].axis("off")

axes[0][2].imshow(cl_chL_lab)
axes[0][2].set_title("L_channel(LAB)")
axes[0][2].axis("off")

axes[0][3].imshow(cl_chY_yuv)
axes[0][3].set_title("Y_channel")
axes[0][3].axis("off")


axes[1][0].imshow(cvt_hsv)
axes[1][0].set_title("V channel")
axes[1][0].axis("off")

axes[1][1].imshow(cvt_hsl)
axes[1][1].set_title("L channel(HSL)")
axes[1][1].axis("off")

axes[1][2].imshow(cvt_lab)
axes[1][2].set_title("L_channel(LAB)")
axes[1][2].axis("off")

axes[1][3].imshow(cvt_yuv)
axes[1][3].set_title("Y_channel")
axes[1][3].axis("off")
```

##**Image Processing(include Gamma)**##

```python
pro_hsv = img_procssing(damage, "HSV")
pro_hsl = img_procssing(damage, "HSL")
pro_lab = img_procssing(damage, "LAB")
pro_yuv = img_procssing(damage, "YUV")

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].imshow(pro_hsv)
axes[0].set_title("HSV")
axes[0].axis("off")

axes[1].imshow(pro_hsl)
axes[1].set_title("HSL")
axes[1].axis("off")

axes[2].imshow(pro_lab)
axes[2].set_title("LAB")
axes[2].axis("off")

axes[3].imshow(pro_yuv)
axes[3].set_title("YUV")
axes[3].axis("off")
```
